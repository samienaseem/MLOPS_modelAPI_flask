```python
import pandas as pd
import numpy as np
import pickle
```

```python
train_df = pd.read_csv('train_lpp.csv') 
train_df.head()
```

```python
train_df.info()
```

## Binary Encoding of Categorical Variables

```python
train_df['Gender']= train_df['Gender'].map({'Male':0, 'Female':1})
train_df['Married']= train_df['Married'].map({'No':0, 'Yes':1})
train_df['Loan_Status']= train_df['Loan_Status'].map({'N':0, 'Y':1})
```

```python
train_df.info()
```

## Checking for Missing Values

```python
train_df.isnull().sum()
```

```python
## dropping all the missing values
train_df = train_df.dropna()
train_df.isnull().sum()
```

## Segregating the target variable from the features

```python
X = train_df[['Gender', 'Married', 'ApplicantIncome', 'LoanAmount', 'Credit_History']]
y = train_df.Loan_Status
X.shape, y.shape
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=5)
```

## Model Training

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(max_depth=4, random_state=5)
model.fit(X_train, y_train)
```

## Cross Validation

```python
from sklearn.metrics import accuracy_score

pred_val = model.predict(X_val)
accuracy_score(y_val, pred_val)
```

```python
pred_train = model.predict(X_train)
accuracy_score(y_train, pred_train)
```

## Saving the model

```python
# saving the model 
import pickle 

pickle_out = open("classifier1.pkl", mode = "wb") 
pickle.dump(model, pickle_out)
pickle_out.close()
```
